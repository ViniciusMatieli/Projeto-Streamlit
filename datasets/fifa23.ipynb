```python
import pandas as pd
```

```python
df = pd.read_csv('FIFA23.csv', delimiter=',',index_col=0)
df.head(4)
```

```python
df['Value']
```

```python
df['Wage'].values
```

```python
df['Wage'] = df['Wage'].replace('','€')
```

```python
df['Wage'].values
```

```python
# Função para converter valores da coluna
def convert_wage(value):
    if pd.isna(value):
        return None
    value = value.replace('€', '')
    try:
        if 'M' in value:
            return float(value.replace('M', '')) * 1e6
        elif 'K' in value:
            return float(value.replace('K', '')) * 1e3
        else:
            return float(value)
    except ValueError:
        return None

# Aplicar a função de conversão à coluna 'Wage'
df['Wage'] = df['Wage'].apply(convert_wage)
df['Value'] = df['Value'].apply(convert_wage)
df['Release Clause'] = df['Release Clause'].apply(convert_wage)
```
