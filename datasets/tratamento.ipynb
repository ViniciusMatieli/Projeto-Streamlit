```python
import pandas as pd
```

```python
df = pd.read_csv('FIFA23_official_data.csv',delimiter=',', index_col=0)
```

```python
df.head(3)
```

## Tratando os dados

```python
# Extraindo o texto após o '>'
df['Position'] = df['Position'].str.extract(r'>\s*(.*)')
```

```python
# Remover números no início da string e espaços que podem vir após os números
df['Name'] = df['Name'].str.replace(r'^\d+\s*', '', regex=True)
```

```python
df[df['Name'] == 'Puyol']
```

```python
#Novo DataFrame
new_df = df[df['Contract Valid Until'] > str(2022)]
new_df.head(3)
```

```python
#Transformando dtype float para int
new_df['Kit Number'] = new_df['Kit Number'].astype(str)
```

```python
new_df.to_csv('FIFA23.csv')
```
